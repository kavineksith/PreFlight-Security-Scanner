## Step 4: View Reports

The scanner generates reports in HTML, JSON, and CSV formats in the `scan_reports` directory.

```python
import os
from IPython.display import HTML, display

reports_dir = './scan_reports'
if os.path.exists(reports_dir):
    html_files = [f for f in os.listdir(reports_dir) if f.endswith('.html')]
    if html_files:
        # Sort to get the latest report
        latest_html = sorted(html_files)[-1]
        print(f"Found report: {latest_html}")
        
        # To download it to your local machine:
        from google.colab import files
        files.download(os.path.join(reports_dir, latest_html))
    else:
        print("No HTML reports found.")
else:
    print("scan_reports directory not found. Did the scanner run successfully?")
```
